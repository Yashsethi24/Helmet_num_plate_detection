Yolo Model Training Code

| Parameter               | Description                                                      |
|--------------------------|------------------------------------------------------------------|
| `task=detect`            | Sets the task type to object detection for helmet detection.     |
| `mode=train`             | Specifies the training mode to train the model with labeled data.|
| `model=yolov8l.pt`       | Uses the pre-trained YOLOv8 large model for better accuracy.     |
| `data=.../data.yaml`     | Path to the dataset configuration file with class names and image paths.|
| `epochs=120`             | Number of training epochs, providing enough learning time while avoiding overfitting.|
| `imgsz=640`              | Input image size of 640x640 pixels, balancing accuracy and speed.|
| `batch=16`               | Number of images processed per batch, optimized for GPU memory usage.|
| `device=0`               | Specifies the first GPU for faster training; use `device=cpu` if no GPU is available.|
| `workers=4`              | Number of data loading workers to speed up image preprocessing.  |
| `optimizer=SGD`          | Uses Stochastic Gradient Descent for stable and consistent learning.|
| `lr0=0.001`              | Initial learning rate, balancing fast learning with stability.   |
| `lrf=0.1`                | Final learning rate fraction, allowing gradual learning rate decay.|
| `weight_decay=0.0005`    | L2 regularization to prevent overfitting by penalizing large weights.|
| `momentum=0.937`         | Momentum for the SGD optimizer to maintain gradient direction.   |
| `patience=20`            | Early stopping patience; stops training if no improvement for 20 epochs.|
| `project=.../helmet_detection` | Directory to save training results, including models, logs, and metrics.|
| `name=helmet_detection_balanced` | Experiment name for easy tracking of training runs.     |
| `cache=True`             | Caches images to accelerate training by reducing disk I/O.       |
| `save_period=10`         | Saves model checkpoints every 10 epochs for safety and monitoring.|
| `exist_ok=True`          | Allows reusing the same directory without overwriting existing files.|
| `augment=True`           | Enables data augmentation to improve generalization.             |
| `conf=0.5`               | Confidence threshold; predictions below 50% are discarded to reduce false positives.|
| `iou=0.5`                | Intersection over Union threshold for non-max suppression, balancing precision and recall.|
| `mosaic=0.3`             | Combines 4 images into one for improved context learning with minimal distortion.|
| `mixup=0.1`              | Blends two images slightly to enhance robustness against occlusions.|
| `hsv_h=0.01`             | Small hue variation of Â±1% for natural color augmentation.       |
| `hsv_s=0.3`              | Saturation variation of Â±30% to simulate different lighting conditions.|
| `hsv_v=0.2`              | Brightness variation of Â±20% for moderate lighting changes.      |
| `flipud=0.0`             | No vertical flipping, as it is unnatural for helmet detection scenarios.|
| `fliplr=0.3`             | 30% chance of horizontal flipping to enhance spatial robustness. |
| `degrees=3.0`            | Allows minimal rotation to handle slight tilts in object orientation.|
| `translate=0.05`         | Translates images by Â±5% of size, simulating small camera movements.|
| `scale=0.1`              | Scales images by Â±10% to improve robustness against size variations.|
| `shear=0.05`             | Shears images by Â±5% for enhanced geometric robustness.          |
| `perspective=0.0`        | No perspective transformation to maintain natural viewing angles.|
| `amp=True`               | Enables Automatic Mixed Precision for faster training and reduced GPU memory usage.|
| `agnostic_nms=False`     | Keeps Non-Max Suppression class-specific, ensuring `helmet` and `no_helmet` detections remain distinct.|


```python
!yolo task=detect mode=train model=yolov8l.pt data=../content/drive/MyDrive/Datasets/BikeHelmet/data.yaml epochs=120 imgsz=640 batch=16 device=0 workers=4 optimizer=SGD lr0=0.001 lrf=0.1 weight_decay=0.0005 momentum=0.937 patience=20 project=runs/detect/helmet_detection name=helmet_detection_balanced cache=True save_period=10 exist_ok=True augment=True conf=0.5 iou=0.5 mosaic=0.3 mixup=0.1 hsv_h=0.01 hsv_s=0.3 hsv_v=0.2 flipud=0.0 fliplr=0.3 degrees=3.0 translate=0.05 scale=0.1 shear=0.05 perspective=0.0 amp=True agnostic_nms=False
```
